```python
def generator_block(in_Channels, out_channels, stride, kernel_size,padding, bias = False, last_layer = False):
    if not last_layer:
        return nn.Sequential(
            nn.ConvTranspose2d(in_Channels, out_channels, 
                               kernel_size=kernel_size, stride=stride, padding=padding, bias=bias),
            nn.BatchNorm2d(out_channels),
            nn.ReLU(True),
        )
    else:
        return nn.Sequential(
            nn.ConvTranspose2d(in_Channels, out_channels, kernel_size=kernel_size, 
                               stride=stride, padding=padding, bias=bias),
            nn.Tanh()
        )
```

```python
class Generator(nn.Module):
    def __init__(self, input_dim = 128, hidden_dim = 512, output_dim = 3):
        super(Generator, self).__init__()
        self.gen = nn.Sequential(
            generator_block(input_dim, hidden_dim, kernel_size = 4, stride = 1, padding = 0),
            generator_block(hidden_dim, hidden_dim//2, kernel_size = 4, stride = 2, padding = 1),
            generator_block(hidden_dim//2, hidden_dim//4, kernel_size = 4, stride = 2, padding = 1),
            generator_block(hidden_dim//4, hidden_dim//8, kernel_size = 4, stride = 2, padding = 1),
            generator_block(hidden_dim//8, output_dim, 
                                 kernel_size = 4, stride = 2, padding = 1, last_layer = True)
        )
        
        

    def forward(self, x):
        return self.gen(x)
        
        
```

```python
def discriminator_block(in_Channels, out_channels, stride, 
                        kernel_size,padding, bias = False, last_layer = False):
    if not last_layer:
        return nn.Sequential(
            nn.Conv2d(in_Channels, out_channels, kernel_size=kernel_size, 
                      stride=stride, padding=padding, bias=False),
            nn.BatchNorm2d(out_channels),
            nn.LeakyReLU(0.2, inplace=True),
        )
    else:
        return nn.Sequential(
            nn.Conv2d(in_Channels, out_channels, kernel_size=kernel_size, 
                      stride=stride, padding=padding, bias=False),
            nn.Flatten(),
            nn.Linear(out_channels, 1)
        )
    
```

```python
class Discriminator(nn.Module):
    def __init__(self, input_dim=3, hidden_dim=64, output_dim=1):
        super(Discriminator, self).__init__()
        self.disc = nn.Sequential(
            discriminator_block(input_dim, hidden_dim, kernel_size=4, stride=2, padding=1, bias=False),
            discriminator_block(hidden_dim, hidden_dim*2, kernel_size=4, stride=2, padding=1, bias=False),
            discriminator_block(hidden_dim*2, hidden_dim*4, kernel_size=4, stride=2, 
                                padding=1, bias=False),
            discriminator_block(hidden_dim*4, hidden_dim*8, kernel_size=4, stride=2, 
                                padding=1, bias=False),
            discriminator_block(hidden_dim*8, output_dim, kernel_size=4, stride=1, 
                                padding=0, bias=False, last_layer = True)
        )
        
    def forward(self, x):
        return self.disc(x)
```

```python
def randn(batch_size, latent_size, dim, device='cuda'):
    return torch.randn(batch_size, latent_size, dim, dim, device = device)
    
```

```python
def disc_loss(gen, disc, real, criterion, batch_size, latent_size, device):
    #gen 
    random_iamges = randn(batch_size, latent_size, 1, device = device)
    generated_images = gen(random_iamges)
    generated_images_pred = disc(generated_images.detach())
    generated_loss = criterion(generated_images_pred, torch.zeros_like(generated_images_pred))
    
    #disc
    real_pred = disc(real)
    real_loss = criterion(real_pred, torch.ones_like(real_pred))
    
    return (real_loss + generated_loss)/2
```

```python
def gen_loss(gen, disc, criterion, batch_size, latent_size, device):
    random_iamges = randn(batch_size, latent_size, 1, device = device)
    generated_images = gen(random_iamges)
    generated_images_pred = disc(generated_images)
    loss = criterion(generated_images_pred, torch.ones_like(generated_images_pred))
    return loss
```

```python
from tqdm import tqdm
def Train(gen, disc, iterator, criterion, gen_optimizer, disc_optimizer):
    gen.train()
    disc.train()
    mean_discriminator_loss = 0
    mean_generator_loss = 0
    for image, _ in tqdm((iterator), ncols = 75, desc ="Train", total = len(iterator)):
        len_size = len(image.shape)
        image = image.to(device)
        #generator
        gen_optimizer.zero_grad()
        gen_loss_1 = gen_loss(gen, disc, criterion, len_size, 128, device)
        gen_loss_1.backward()
        gen_optimizer.step()
        mean_generator_loss = gen_loss_1.item()/len(iterator)
        #disc optimizer
        disc_optimizer.zero_grad()
        disc_loss_1 = disc_loss(gen, disc, image, criterion, len_size, 128, device)
        disc_loss_1.backward()
        disc_optimizer.step()
        mean_discriminator_loss = disc_loss_1.item()/len(iterator)
        
    return mean_generator_loss, mean_discriminator_loss
        
        
        
    
```
